```python
# while문 연습 : 나무찍기

tree_hit = 1
while tree_hit < 10:
    print(f'나무를 {tree_hit}번 찍었습니다.')
    tree_hit += 1
print("나무가 넘어갑니다.")      
```

```python
# while문 연습 : 입력값에 따라 반복 종료하기
# 사용자로부터 숫자를 하나씩 입력받아서 총합과 평균을 출력
# 사용자가 0을 입력하면 숫자 입력을 멈춤

sum = count = 0
num = int(input("숫자 입력 (0 : 끝) :"))
while num != 0:
      sum += num
      count += 1
      num = int(input("숫자 입력 (0 : 끝) :"))
                      
# 0으로 나누기를 하면  zerodivisionerror 발생
if count != 0:                   
    average = sum / count
else:
    average = 0
    
print(f">> 총합 : {sum}, 평균 : {average}")
```

```python
# for문 연습

sum = 0
for i in 1.1, 2.5, 3.6, 4.2, 5.4:
    sum += i
    print(i, sum)
print(f'합 : {sum}, 평균 : {sum / 5:.2f}')
```

```python
# 5개 숫자를 계속 입력받아 모두 더한 값 출력
print('5개의 숫자를 입력하세요.')
sum = 0
for _ in range(5):
    num = int(input())
    sum += num
    
print(f'5개 숫자의 합은 {sum}입니다..')
```

```python
# 1부터 N까지의 합 구하기

number = int(input('숫자를 입력하세요 :'))
sum = 0
for num in range(1, number + 1):
    #print(num)
    sum += num
    
print(f'1부터 {number}까지의 합은 {sum}입니다.')
```
