### 리스트의 항목에 접근하기: indexing, for문

```python
test_list =[1, 2, 3, 'one', 'two', 'three']

print(test_list[0])
print(test_list[3])
print(test_list[-1])
print(test_list[-3])

for item in test_list:
    print(item)
```

```python
test_list =[11, 22, 33, 'tiger', 'monkey', 'ostrich']

print(test_list[0])
print(test_list[-3])
print(test_list[-1])
print(test_list[1])

for item in test_list:
    print(item)
```

### 리스트의 항목에 접근하기: 슬라이싱(slicing)

```python
test_list =[1, 2, 3, 4, 5, 6, 7, 8]

print(test_list[:4])
print(test_list[4:])
print(test_list[1::3])
print(test_list[:])
print(test_list[::])
print(test_list[::3])
print(test_list[:5:2])

print(test_list[-1:-7:-2])
print(test_list[1:-1:-1])
print(test_list[-3::-1])
print(test_list[::-1])
print(test_list[::-2])
```

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]
test_list[:2] = [111, 222, 333]
print(test_list)
```

```python
print([1, 2, 3][::-1])
```

### 리스트 항목 재배열 하기

```python
test_list.reverse()
print(test_list)
```

```python
a_list = ['one', 'two', 'three', 'four']
print('원본 데이터', a_list)
a_list.reverse()
print('역순 후 데이터', a_list)

a_list.sort()
print('역순 후 데이터', a_list)
a_list = ['one', 'two', 'three', 'four']
b_list = sorted(a_list)
print('원본 데이터', a_list)
print('정렬 후 데이터', b_list)
```

```python
scores = [95, 45, 21, 100, 30]
print(sorted(scores, reverse=True))

print(scores)
scores.sort(reverse=True)
print(scores)
```

```python
# 추가 문제 1
numbers = [3, 1, 4, 1, 5, 9, 2]
print(f"원본 리스트: {numbers}")

numbers.sort()
print(f"정렬 후 (sort): {numbers}")

numbers.reverse()
print(f"순서 뒤집기 (reverse): {numbers}")
```

```python
# 추가 문제 2
fruits = ["banana", "apple", "cherry", "dragonfruit"]
print("원본 리스트", fruits)

sorted_fruits = sorted(fruits)
print("새로운 정렬 리스트", sorted_fruits)
print("원본 유지 확인", fruits)

sorted_fruits.reverse()
print("순서 뒤집기", sorted_fruits)
```

```python
# 1부터 N까지의 수로 이루어진 리스트 생성
N = int(input("숫자 :"))
num_list = []
for i in range(1, N + 1):
    num_list.append(i)
print(num_list)
```

```python
# 1부터 N까지의 수로 이루어진 리스트 생성
N = int(input("숫자 :"))
num_list = [i for i in range(1, N + 1)]
print(num_list)

#expression
num_list = [i * 2 for i in range(1, N + 1)]
print(num_list)

# if문 사용
N = int(input("숫자 :"))
num_list = [i for i in range(1, N + 1) if i % 2 == 0]
print(num_list)
```
