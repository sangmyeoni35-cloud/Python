```python
# 리스트와 반복 구조
# 주어진 데이터가 어디에 저장되어 있는지 출력

data_list = [2, 42, 56, 13, 78]
target = 56
index = 0

while index <= 5 and data_list[index] != target:
    index += 1
print(f'>> index : {index}')
```

```python
# 리스트에서 가장 큰 값 찾기

data_list = [2, 42, 56, 13, 78]

max = data_list[0]
max_index = 0

for i in range(5):
    if max < data_list[i]:
        max = data_list[i]
        
    max_index = i

print(f'>> max : {max}, max index : {max_index}')
```

```python
# 리스트에서 가장 큰 값 찾기

data_list = [2, 42, 56, 13, 78]
max = data_list[0]
index = 0
max_index = 0

for data in data_list :
    if max < data :
        max = data
        max_index = index
        index += 1
print(f'>> max : {max}, max_index : {max_index}')
```

```python
# 리스트에서 가장 작은 값 찾기

data_list = [2, 42, 56, 13, 78]

min = data_list[0]
index = 0
min_index = 0

for data in data_list :
    if min > data :
        min = data
        min_index = index
        index += 1

print(f'>> min : {min}, min index : {min_index}')
```

```python
# 리스트에 있는 수들 중 입력값 보다 큰 수가 몇개 있는지

data_list = [2, 42, 56, 13, 78]
num = int(input('숫자 입력 :'))
cnt = 0

for data in data_list:
    if data > num:
        cnt += 1
print(f'입력된 수보다 큰 숫자는 {cnt}개입니다.')
```

```python
# while문 연습 : 나무 찍기

tree_hit = 0
while tree_hit < 10:
    tree_hit += 1
    print(f'나무를 (tree_hit)번 찍었습니다.')

print('나무가 쓰러졌습니다.')
```

```python
# while문 연습 : 입력값에 따라 반복 종료하기
# 사용자로부터 숫자를 하나씩 입력받아서 총합과 평균을 출력
# 사용자가 0을 입력하면 숫자 입력을 멈춤

sum = count = 0
num = int(input("숫자 입력 (0 : 끝) :"))
while num != 0:
      sum += num
      count += 1
      num = int(input("숫자 입력 (0 : 끝) :"))
                      
# 0으로 나누기를 하면  zerodivisionerror 발생
if count != 0:                   
    average = sum / count
else:
    average = 0
    
print(f">> 총합 : {sum}, 평균 : {average}")
```

```python
# 무한 루프와 분기문 break 활용

sum = cnt = 0

while True:
    num = int(input('숫자 입력(0을 입력하면 끝) :'))
    if num == 0 : break
    sum += num
    cnt += 1

if cnt != 0:
    average = float(sum / cnt)
else:
    average = 0

print(f'>> 총합은 {sum}이고, 평균은 {average : .2f}이다.')
```

```python
# 5개 숫자를 계속 입력받아 모두 더한 값 출력
print('5개의 숫자를 입력하세요.')
sum = 0
for _ in range(5):
    num = int(input())
    sum += num
    
print(f'5개 숫자의 합은 {sum}입니다..')
```

```python
# 1부터 N까지의 합 구하기

number = int(input('숫자를 입력하세요 :'))
sum = 0
for num in range(1, number + 1):
    #print(num)
    sum += num
    
print(f'1부터 {number}까지의 합은 {sum}입니다.')
```

```python
# for문 연습 : 숫자를 입력받고 그 숫자에 해당하는 구구단 출력

num = int(input('몇 단을 출력할까? (1~9까지 숫자 입력)'))

if 1 <= num <= 9:
    for i in range(1,10):
        print(f'{num} * (i) = {num * i}')

else:
    print('범위 밖이라서 구구단을 출력할 수 없어!')
    
```

```python
# FOR문 연습 : 숫자를 입력받고 그 수의 팩토리얼을 출력

fac = 1
num = int(input(' 몇 팩토리얼을 출력할까?:'))

for i in range(2, num + 1):
    fac += i

print(f'{num}! = {fac}')
```
