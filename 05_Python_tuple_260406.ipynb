```python
### 튜플 다루기

# 튜플 항목 접근 - 인덱싱, 슬라이싱, for ~ in
t = (10, 20, 30, 40, 50)
print(t[0]) # 인덱싱
print(t[1:4]) # 슬라이싱
for i in t:   # 반복문 
    print(i, end=' ')

# 튜플 탐색 - in 연산자, index(), count
print(30 in t)
print(t. index(40))
print(t. count(20))

# 튜플 관련 정보 - len(), max(), min(), sum()
print(len(t))
print(max(t))
print(min(t))
print(sum(t))

# 새로운 튜플 생성
t2 = sorted(t)
print(tuple(t2))
t3 = t + (60, 70)
print(t3)

# 튜플삭제
del t4
```
