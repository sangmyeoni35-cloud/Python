```python
### 리스트와 반복구조 연습
def getIndex(lst, value):
    for i, v in enumerate(lst):
        if v == value:
            return i
    return -1  # 없으면 -1 반환

def getMax(lst):
    max_val = lst[0]
    for v in lst:
        if v > max_val:
            max_val = v
    return max_val
number_list = [23, 45, 27, 11, 25, 65, 78]

print(getIndex(number_list, 25))
print(getMax(number_list))

def countGT(lst, threshold):
    count = 0
    for v in lst:
        if v > threshold:
            count += 1
    return count

number_list = [23, 45, 27, 11, 25, 65, 78]

print(countGT(number_list, 42))
def sumList(lst):
    total = 0
    for v in lst:
        total += v
    return total

number_list = [23, 45, 27, 11, 25, 65, 78]

print(sumList(number_list)) 

def swapList(lst):
    return lst[::-1]

number_list = [23, 45, 27, 11, 25, 65, 78]
number_list = swapList(number_list)
print(number_list) 
```
